```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
X=np.array([1.2,1.4,1.9,2.3,2.8,3.2])
Y=np.array([1.9,2.1,2.5,2.7,3.0,3.4])
```

```
plt.scatter(X,Y)
plt.grid()
plt.xlabel("weight")
plt.ylabel("Price")
```

```
theta=np.array([[0],[0.1]]) #initialising cost function
def hypothesis(x,theta):
  h=theta[0]+theta[1]*x
  return h
```

```
def cost(X,Y,theta):
  m=len(X)
  error=0
  for i in range(m):
    h=hypothesis(X[i],theta)
    error=error+((h-Y[i])**2)
  error=error/(2*m)
  return error
```

```
def gradient(X,Y,theta):
  grad=np.zeros(2)
  m=len(X)
  
  for i in range(m):
    x=X[i]
    y=Y[i]
    h=hypothesis(x,theta)
    grad[0]+=2*(h-y)
    grad[1]+=2*(h-y)*x

    return grad
```

```
def gradientdesc(X,Y,lr,epoch):
  m=len(X)
  for i in range(epoch):
    grad=gradient(X,Y,theta)
    ssi=grad[0]*lr
    sss=grad[1]*lr
    theta[0]=theta[0]-ssi
    theta[1]=theta[1]-sss
  return theta  
  
```

```
th=gradientdesc(X,Y,0.1,1000)
print(th)
error=cost(X,Y,theta)
print(error)
```

```
l=[]
k=[]
for i in range(50,1000,100):
  k.append(i)
  theta=gradientdesc(X,Y,0.01,i)
  l.append(theta)
print(k)
p=[]
for j in range(len(l)):
  error=cost(X,Y,l[j])
  m=np.sum(error)
  p.append(m)
print(p)
```
